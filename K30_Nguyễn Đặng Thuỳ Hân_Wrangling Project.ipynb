###5. Using transactions.csv
Define type of transactions (‘transaction_type’) for each row, given:
- transType = 2 & merchant_id = 1205: Bank Transfer Transaction
- transType = 2 & merchant_id = 2260: Withdraw Money Transaction
- transType = 2 & merchant_id = 2270: Top Up Money Transaction
- transType = 2 & others merchant_id: Payment Transaction
- transType = 8, merchant_id = 2250: Transfer Money Transaction
- transType = 8 & others merchant_id: Split Bill Transaction
- Remained cases are invalid transactions


```python
transaction_cleaned1.head(10)
```

```python
transaction_cleaned1.dtypes
```

```python
conditions = [(transaction_cleaned1['transType']==2.0) & (transaction_cleaned1['merchant_id'].str.contains('1205')),
              (transaction_cleaned1['transType']==2.0) & (transaction_cleaned1['merchant_id'].str.contains('2260')),
              (transaction_cleaned1['transType']==2.0) & (transaction_cleaned1['merchant_id'].str.contains('2270')),
              (transaction_cleaned1['transType']==2.0) & (~transaction_cleaned1['merchant_id'].isin(['1205','2260','2270'])),
              (transaction_cleaned1['transType']==8.0) & (transaction_cleaned1['merchant_id'].str.contains('2250')),
              (transaction_cleaned1['transType']==2.0) & (~transaction_cleaned1['merchant_id'].str.contains('2250'))]
```

```python
transaction_types =['Bank Transfer Transaction',
                    'Withdraw Money Transaction',
                    'Top Up Money Transaction',
                    'Payment Transaction',
                    'Transfer Money Transaction',
                    'Split Bill Transaction']
```

```python
import numpy as np
```

```python
transaction_cleaned1['transaction_type'] = np.select(conditions,transaction_types, default = 'invalid transactions')
```

```python
transaction_cleaned1['transaction_type'].unique()
```

###6. Of each transaction type (excluding invalid transactions): find the number of transactions, volume, senders and receivers.

```python
valid_transactions = transaction_cleaned1[transaction_cleaned1['transaction_type']!='invalid transactions']
valid_transactions.head()
```

```python
result = valid_transactions.groupby('transaction_type').agg(num_transactions=('transaction_id', 'count'),total_volume =('volume','sum'),nunique_sender=('sender_id','nunique'),unique_receiver=('receiver_id','nunique')).reset_index()
```
